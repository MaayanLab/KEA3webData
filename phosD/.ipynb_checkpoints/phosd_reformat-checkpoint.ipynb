```python
import pandas as pd
import fileinput
```

```python
df = pd.read_csv('input/Supplementary Table S1.txt', delimiter = '\t')
#df = df[df.Score >= .1]
df = df[df.Score >= .5]
```

```python
kin = [i for i in list(df.loc[:, 'Kinase'])]
sub = [i for i in list(df.loc[:, 'Substrate'])]
```

```python
[print('{0}\t{1}'.format(k, s)) for k, s in zip(kin, sub)]
```

```python
### CONVERT FROM UNIPROT ID TO GENE ID

# UNIPROT ID TO GENE ID
df1 = pd.read_csv('input/uniprot_to_gene_hgnc.txt', sep = '\t')
df1 = df1.dropna()
df1.head()
```

```python
toGeneID = {} # maps uniprotID to approved symbol(s)
for x, y in zip(list(df1.loc[:, 'Approved symbol']), list(df1.loc[:, 'UniProt ID(supplied by UniProt)'])):
    if y not in toGeneID:
        toGeneID[y] = {x}
    else:
        toGeneID[y].add(x)
```

```python
# find KSIs
d = {}
for k, s in zip(kin, sub):
    if k in toGeneID and s in toGeneID:
        k = toGeneID[k]
        s = toGeneID[s]
        for k1 in k:
            if k1 not in d:
                d[k1] = {s1 for s1 in list(s)}
            else:
                {d[k1].add(s1) for s1 in list(s)}
```

```python
allnp = open('output_phosd_all_human.gmt', 'w+')
fnp = open('output_phosd_fourplusinteractions_human.gmt', 'w+')
uSubs = set()
numKSI = 0
numKins = 0
for k in d:
    allnp.write('{0}_phosd_human\t'.format(k) + '\t'.join(d[k]) + '\n')
    if len(d[k]) >= 4:
        numKSI += len(d[k])
        numKins += 1
        {uSubs.add(x) for x in d[k]}
        fnp.write('{0}_phosd_human\t'.format(k) + '\t'.join(d[k]) + '\n')
    
        
print('{0}\t#kins: {1}\t#ksi: {2}\t#usubs: {3}'.format('human', numKins, numKSI, len(uSubs)))

allnp.close()
fnp.close()

```
