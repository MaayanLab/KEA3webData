```python
import pandas as pd
import mygene
import fileinput

mg = mygene.MyGeneInfo()
```

```python
df = pd.read_csv('input/reactome.homo_sapiens.interactions.tab-delimited.txt', delimiter = '\t')
#df = pd.read_csv('input/testinp.txt', delimiter = '\t')
prot1 = [i.split(':')[1] for i in list(df.iloc[:, 0])]
prot2 = [i.split(':')[1] for i in list(df.iloc[:, 3])]
```

```python
out1 = mg.querymany(prot1, scopes='uniprot', fields='symbol', species='human')
out2 = mg.querymany(prot2, scopes='uniprot', fields='symbol', species='human')

```

```python
inp = fileinput.FileInput('input/kinases_human.txt')
pkin = []
for line in inp:
    pkin.append(line.split('\n')[0])
```

```python
d = {}
for p1, p2 in zip(out1, out2):
    if 'notfound' not in p1 and 'notfound' not in p2:
        if p1['symbol'] in pkin:
            if p1['symbol'] not in d:
                d[p1['symbol']] = {p2['symbol']}
            else:
                d[p1['symbol']].add(p2['symbol'])
        if p2['symbol'] in pkin:
            if p2['symbol'] not in d:
                d[p2['symbol']] = {p1['symbol']}
            else:
                d[p2['symbol']].add(p1['symbol'])
                
allnp = open('output_reactome_all_human.gmt', 'w+')
fnp = open('output_reactome_fourplusinterations_human.gmt', 'w+')
uSubs = set()
numKSI = 0
for k in d:
    temp = '{0}_reactome_human'.format(k)
    for i in d[k]:
        numKSI += 1
        uSubs.add(i)
        temp += '\t{0}'.format(i)
    allnp.write('{0}\n'.format(temp))
    if len(d[k]) >= 4:
        fnp.write('{0}\n'.format(temp))
        
print('{0}\n#kin: {1}\t#ksi: {2}\t#usubs: {3}'.format('human', len(d), numKSI, len(uSubs)))
```
