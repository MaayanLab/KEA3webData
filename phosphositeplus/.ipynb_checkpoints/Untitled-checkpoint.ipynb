```python
import pandas as pd
import csv
```

```python
## WHEN I RUN THIS DATASET, output files disappear seemingly at random
## am i somehow modifying the data?
## I created this new notebook bc in the old one, two rat files would disappear when I ran everything in a loop





df = pd.read_csv('input/Kinase_Substrate_Dataset EDITED', delimiter = '\t')
df = df[df.KIN_ORGANISM == df.SUB_ORGANISM]
df = df.loc[:, ['KIN_ORGANISM', 'SUB_ORGANISM', 'GENE', 'SUB_GENE', 'SUB_MOD_RSD']]
df['SUB_PHOS'] = df['SUB_GENE'] + '_' + df['SUB_MOD_RSD']

org = list(df.loc[:, 'KIN_ORGANISM'])
gene1 = list(df.loc[:, 'GENE']) # kinase gene
gene2 = list(df.loc[:, 'SUB_GENE']) # substrate gene
psites = list(df.loc[:, 'SUB_PHOS'])

df.head()
```

```python
# i need to run this twice to get all of the output to load?
df_human = df[df.KIN_ORGANISM == 'human'] ##
df_human ##
o = 'human'

d = {}
for k, s_p in zip(df_human.loc[:, 'GENE'], df_human.loc[:, 'SUB_PHOS']): ##
    if k not in d:
        d[k] = {s_p}
    else:
        d[k].add(s_p)

allwp = open('output-withphosphosite_phosphositeplus_all_{0}.gmt'.format(o), 'w+') # w phosphosite
fwp = open('output-withphosphosite_phosphositeplus_fourplusinteractions_{0}.gmt'.format(o), 'w+')
allnp = open('output_phosphositeplus_all_{0}.gmt'.format(o), 'w+')
fnp = open('output_phosphositeplus_fourplusinteractions_{0}.gmt'.format(o), 'w+')
'''allwp = open('output-withphos_{0}.gmt'.format(o), 'w+')
fwp = open('output-withphos_4+_{0}.gmt'.format(o), 'w+')
allnp = open('output_{0}.gmt'.format(o), 'w+')
fnp = open('output_4+_{0}.gmt'.format(o), 'w+')'''
wr1 = csv.writer(allwp, delimiter = '\t')
wr1a = csv.writer(fwp, delimiter = '\t')
wr2 = csv.writer(allnp, delimiter = '\t')
wr2a = csv.writer(fnp, delimiter = '\t')

numKSI = 0
uSubs = set()
for k in d:
    # four+ KSI for this kinase
    if len(d[k]) >= 4: 
        numKSI += len(d[k])
        {uSubs.add(x) for x in d[k]}
        
        temp = list(d[k])
        subs = list({str(x).split('_')[0] for x in temp})
        temp.insert(0, '{0}_phosphositeplus_{1}'.format(k, o))
        subs.insert(0, '{0}_phosphositeplus_{1}'.format(k, o))
        
        # modify _four
        wr1a.writerow(temp)
        wr2a.writerow(subs)
        
        # modify _all
        wr1.writerow(temp)
        wr2.writerow(subs)
    
    # fewer than four KSI for this kinase
    else:
        temp = list(d[k])
        subs = list({str(x).split('_')[0] for x in temp})
        temp.insert(0, '{0}_phosphositeplus_{1}'.format(k, o))
        subs.insert(0, '{0}_phosphositeplus_{1}'.format(k, o))
        
        wr1.writerow(temp)
        wr2.writerow(subs)

        
print('{0}\n#kin: {1}\t#ksi: {2}\t#usubs: {3}'.format(o, len(d), numKSI, len(uSubs)))
   
```
